```python
import pandas as pd
df_customer = pd.read_csv('/content/Dataset.csv')
display(df_customer.head())
```

```python
# Display basic information about the DataFrame
display(df_customer.info())
```

```python
# Check for missing values
display(df_customer.isnull().sum())
```

```python
# Check for duplicate rows
print(f"Number of duplicate rows: {df_customer.duplicated().sum()}")

# Remove duplicate rows if any
df_customer.drop_duplicates(inplace=True)
print(f"Number of rows after removing duplicates: {len(df_customer)}")
```

```python
# Impute missing 'Review Rating' values with the median
median_review_rating = df_customer['Review Rating'].median()
df_customer['Review Rating'].fillna(median_review_rating, inplace=True)

print(f"Missing values in 'Review Rating' after imputation: {df_customer['Review Rating'].isnull().sum()}")
```

```python
# Check unique values for object columns
for column in df_customer.select_dtypes(include='object').columns:
    print(f"Unique values for '{column}':\n{df_customer[column].unique()}\n")

# Display descriptive statistics for numerical columns to identify outliers
display(df_customer.describe())
```

```python
# Calculate total spend and purchase frequency per customer
customer_value_metrics = df_customer.groupby('Customer ID').agg(
    total_spend=('Purchase Amount (USD)', 'sum'),
    purchase_frequency=('Customer ID', 'count')
).reset_index()

# Calculate Average Order Value (AOV)
customer_value_metrics['avg_order_value'] = customer_value_metrics['total_spend'] / customer_value_metrics['purchase_frequency']

# Display the first few rows of the customer value metrics
display(customer_value_metrics.head())
```

```python
# Calculate discounted spend and promo code usage per customer
promo_metrics = df_customer.groupby('Customer ID').agg(
    discounted_spend=('Purchase Amount (USD)', lambda x: x[df_customer.loc[x.index, 'Discount Applied'] == 'Yes'].sum()),
    promo_code_orders=('Promo Code Used', lambda x: (x == 'Yes').sum())
).reset_index()

# Merge with customer_value_metrics to get total_spend and purchase_frequency
promo_metrics = pd.merge(promo_metrics, customer_value_metrics[['Customer ID', 'total_spend', 'purchase_frequency']], on='Customer ID')

# Calculate discount_rate
promo_metrics['discount_rate'] = promo_metrics['discounted_spend'] / promo_metrics['total_spend']
promo_metrics['discount_rate'].fillna(0, inplace=True) # Handle cases where total_spend is 0 or no discounted spend

# Calculate promo_order_ratio
promo_metrics['promo_order_ratio'] = promo_metrics['promo_code_orders'] / promo_metrics['purchase_frequency']
promo_metrics['promo_order_ratio'].fillna(0, inplace=True) # Handle cases where purchase_frequency is 0 or no promo code orders

# Display the first few rows of the promo dependency metrics
display(promo_metrics.head())
```

```python
# Combine all customer-level metrics into a single DataFrame
customer_features = pd.merge(customer_value_metrics, promo_metrics[['Customer ID', 'discounted_spend', 'promo_code_orders', 'discount_rate', 'promo_order_ratio']], on='Customer ID')

# Define loyalty based on purchase frequency
customer_features['frequency_loyalty_score'] = customer_features['purchase_frequency']

# Check number of unique values in 'frequency_loyalty_score'
unique_frequencies = customer_features['frequency_loyalty_score'].nunique()

if unique_frequencies == 1:
    # If all values are the same, assign a single category
    customer_features['frequency_loyalty_segment'] = 'Single Frequency'
else:
    # If there are enough unique values, apply qcut
    # Adjust q to be at most the number of unique values if duplicates='drop'
    q_val = min(4, unique_frequencies)
    # Define labels based on the actual number of quantiles
    labels = ['Low Frequency', 'Medium Frequency', 'High Frequency', 'Very High Frequency'][:q_val]
    customer_features['frequency_loyalty_segment'] = pd.qcut(customer_features['frequency_loyalty_score'], q=q_val, labels=labels, duplicates='drop')

# Display the distribution of frequency loyalty segments
display(customer_features[['Customer ID', 'purchase_frequency', 'frequency_loyalty_score', 'frequency_loyalty_segment']].head())
print("\nFrequency Loyalty Segment Distribution:")
print(customer_features['frequency_loyalty_segment'].value_counts())
```

```python
# Define loyalty based on total spend
customer_features['revenue_loyalty_score'] = customer_features['total_spend']

# Create revenue loyalty segments (e.g., using quartiles)
# Check number of unique values in 'revenue_loyalty_score'
unique_spend = customer_features['revenue_loyalty_score'].nunique()

if unique_spend == 1:
    # If all values are the same, assign a single category
    customer_features['revenue_loyalty_segment'] = 'Single Revenue'
else:
    # If there are enough unique values, apply qcut
    # Adjust q to be at most the number of unique values if duplicates='drop'
    q_val_revenue = min(4, unique_spend)
    # Define labels based on the actual number of quantiles
    labels_revenue = ['Low Revenue', 'Medium Revenue', 'High Revenue', 'Very High Revenue'][:q_val_revenue]
    customer_features['revenue_loyalty_segment'] = pd.qcut(customer_features['revenue_loyalty_score'], q=q_val_revenue, labels=labels_revenue, duplicates='drop')

# Display the distribution of revenue loyalty segments
display(customer_features[['Customer ID', 'total_spend', 'revenue_loyalty_score', 'revenue_loyalty_segment']].head())
print("\nRevenue Loyalty Segment Distribution:")
print(customer_features['revenue_loyalty_segment'].value_counts())

# Merge 'Review Rating' into customer_features from df_customer for analysis
customer_features = pd.merge(customer_features, df_customer[['Customer ID', 'Review Rating']], on='Customer ID', how='left')
```

```python
# Analyze revenue-based loyalty segments
revenue_loyalty_analysis = customer_features.groupby('revenue_loyalty_segment').agg(
    Avg_Total_Spend=('total_spend', 'mean'),
    Avg_AOV=('avg_order_value', 'mean'),
    Avg_Discount_Rate=('discount_rate', 'mean'),
    Avg_Promo_Order_Ratio=('promo_order_ratio', 'mean'),
    Avg_Review_Rating=('Review Rating', 'mean')
).reset_index()

display(revenue_loyalty_analysis.sort_values(by='Avg_Total_Spend', ascending=False))
```

```python
display(customer_features[['Customer ID', 'total_spend', 'revenue_loyalty_segment']].head())
print("\nRevenue-Based Loyalty Segments (Value Tiers):")
print(customer_features['revenue_loyalty_segment'].value_counts())
```

```python
# Check unique values in 'promo_order_ratio' and categorize manually if few unique values
unique_promo_ratios = customer_features['promo_order_ratio'].nunique()

if unique_promo_ratios <= 2:
    # Assuming the two most common values are 0.0 and 1.0 (or similar)
    # Categorize explicitly if there are only a few distinct values
    customer_features['promo_dependency_segment'] = customer_features['promo_order_ratio'].apply(lambda x: 'High Promo Dependency' if x > 0 else 'No Promo Dependency')
else:
    # If there are enough unique values, apply qcut
    q_val_promo = min(4, unique_promo_ratios)
    labels_promo = ['Low Dependency', 'Medium Dependency', 'High Dependency', 'Very High Dependency'][:q_val_promo]
    customer_features['promo_dependency_segment'] = pd.qcut(customer_features['promo_order_ratio'], q=q_val_promo, labels=labels_promo, duplicates='drop')

display(customer_features[['Customer ID', 'promo_order_ratio', 'promo_dependency_segment']].head())
print("\nPromo Dependency Segment Distribution:")
print(customer_features['promo_dependency_segment'].value_counts())
```

```python
# Create combined segment
customer_features['combined_segment'] = customer_features['revenue_loyalty_segment'].astype(str) + ' - ' + customer_features['promo_dependency_segment'].astype(str)

display(customer_features[['Customer ID', 'revenue_loyalty_segment', 'promo_dependency_segment', 'combined_segment']].head())
print("\nCombined Segment Distribution:")
print(customer_features['combined_segment'].value_counts())
```

```python
# Define 'ideal customer' flag
customer_features['is_ideal_customer'] = ((customer_features['revenue_loyalty_segment'] == 'Very High Revenue') &
                                         (customer_features['promo_dependency_segment'] == 'No Promo Dependency'))

# Display ideal customer distribution
display(customer_features[['Customer ID', 'revenue_loyalty_segment', 'promo_dependency_segment', 'is_ideal_customer']].head())
print("\nIdeal Customer Distribution:")
print(customer_features['is_ideal_customer'].value_counts())
```

```python
# Save the customer_features DataFrame to a CSV file
customer_features.to_csv('customer_features.csv', index=False)
print("Customer features saved to 'customer_features.csv'")

# Feature Dictionary
feature_dictionary = {
    'Customer ID': 'Unique identifier for each customer.',
    'total_spend': 'Total amount of money (USD) spent by the customer across all purchases.',
    'purchase_frequency': 'Total number of purchases made by the customer. (Note: All customers in this dataset have a frequency of 1).',
    'avg_order_value': 'Average amount of money (USD) spent per purchase by the customer.',
    'discounted_spend': 'Total amount of money (USD) spent by the customer on items where a discount was applied.',
    'promo_code_orders': 'Total number of purchases made by the customer where a promo code was used.',
    'discount_rate': 'The proportion of the total spend that was subject to a discount (discounted_spend / total_spend).',
    'promo_order_ratio': 'The ratio of purchases made using a promo code to the total number of purchases (promo_code_orders / purchase_frequency).',
    'Review Rating': 'Average review rating given by the customer across all their purchases.',
    'frequency_loyalty_score': 'Loyalty score based on purchase frequency. (In this dataset, it equals purchase_frequency).',
    'frequency_loyalty_segment': 'Categorical segment based on purchase frequency (e.g., Single Frequency).',
    'revenue_loyalty_score': 'Loyalty score based on total spend. (In this dataset, it equals total_spend).',
    'revenue_loyalty_segment': 'Categorical segment based on total spend (e.g., Low Revenue, High Revenue).',
    'promo_dependency_segment': 'Categorical segment based on reliance on promotions (e.g., No Promo Dependency, High Promo Dependency).',
    'combined_segment': 'A concatenated string combining the revenue loyalty segment and promo dependency segment.',
    'is_ideal_customer': 'Boolean flag indicating if a customer is considered ideal (Very High Revenue and No Promo Dependency).'
}

print("\nFeature Dictionary:")
for feature, description in feature_dictionary.items():
    print(f"- {feature}: {description}")
```
